## Extracting a row

Let's print the number of genes humans share with banana. iloc part informs the computer that we are requesting information about a certain position. The number before the comma is for the row and the one after the comma is for the column. Let's only specify the row of banana:

```
dataset.iloc[1,]
```

You'll notice that the code printed not the row for banana but for the cat. This is because in python indexing begins with 0 instead of 1. So we need to print the 0th row to get the line concerning banana.

```
dataset.iloc[0,]
```

What if I want to print the number for the dog? Please insert the number of the row that reports about the dog within the squared brackets:
